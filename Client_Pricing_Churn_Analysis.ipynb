```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

client_data_path = r'D:\PROJECT\BCG\client_data (1).csv'
price_data_path = r'D:\PROJECT\BCG\price_data (1).csv'

client_data = pd.read_csv(client_data_path)
price_data = pd.read_csv(price_data_path)

print("Client Data"); print(client_data.head())
print("\nClient Data Types"); print(client_data.dtypes)
print("\nClient Data Descriptive Statistics"); print(client_data.describe(include='all'))

numerical_columns_client = client_data.select_dtypes(include=['float64', 'int64']).columns
for column in numerical_columns_client:
    plt.figure(figsize=(10, 5))
    sns.histplot(client_data[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

categorical_columns_client = client_data.select_dtypes(include=['object']).columns
for column in categorical_columns_client:
    plt.figure(figsize=(10, 5))
    sns.countplot(y=client_data[column])
    plt.title(f'Distribution of {column}')
    plt.xlabel('Count')
    plt.ylabel(column)
    plt.show()

print("Price Data"); print(price_data.head())
print("\nPrice Data Types"); print(price_data.dtypes)
print("\nPrice Data Descriptive Statistics"); print(price_data.describe(include='all'))

numerical_columns_price = price_data.select_dtypes(include=['float64', 'int64']).columns
for column in numerical_columns_price:
    plt.figure(figsize=(10, 5))
    sns.histplot(price_data[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

categorical_columns_price = price_data.select_dtypes(include=['object']).columns
for column in categorical_columns_price:
    plt.figure(figsize=(10, 5))
    sns.countplot(y=price_data[column])
    plt.title(f'Distribution of {column}')
    plt.xlabel('Count')
    plt.ylabel(column)
    plt.show()

if 'churn' in client_data.columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(x='churn', data=client_data)
    plt.title('Churn Distribution')
    plt.xlabel('Churn')
    plt.ylabel('Count')
    plt.show()
else:
    print("The churn column is not present in the client data.")

client_data.to_csv('cleaned_client_data.csv', index=False)
price_data.to_csv('cleaned_price_data.csv', index=False)
```
